```python
import matplotlib.pyplot as plt
import seaborn as sb
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
sb.countplot(A.Type)

plt.subplot(2,3,2)
sb.countplot(A.Origin)

plt.subplot(2,3,3)
sb.countplot(A.Cylinders)

plt.subplot(2,3,4)
sb.countplot(A.Passengers)

plt.subplot(2,3,5)
sb.countplot(A.AirBags)

plt.subplot(2,3,6)
sb.countplot(A.DriveTrain)
```

```python
cat = []
con = []
```

```python
for i in A.columns:
    if(A[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

# Create countplot|bar chart for all categorical variables in 1 figure area

```python
cat
```

```python
plt.figure(figsize=(20,20))

x = 1
for i in cat:
    plt.subplot(3,3,x)
    sb.countplot(A[i])
    x = x+1
```

```python
plt.figure(figsize=(20,12))

x = 1
for i in con:
    plt.subplot(3,7,x)
    A[i].hist()
    x = x+1
```

# Bivariate Analysis:

```python
import matplotlib.pyplot as plt
plt.scatter(A.Price,A.Weight)
```

```python
plt.scatter(A.Price,A.Weight)
plt.xlabel("Price of the car")
plt.ylabel("Weight of the car")
plt.title("Price vs Weight scatterplot")
```

```python
for i in A.columns:
    if(A[i].dtypes=="object"):
        sb.countplot(A.Origin,hue=A[i])      #Countplot with hue between categorical and categorical columns
        plt.ylabel(i)
        plt.show()
```
